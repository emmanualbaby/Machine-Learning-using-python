```python
import matplotlib.pyplot as plt
import numpy as np

# Define specific X and Y coordinates
x = np.linspace(0, 10, 100)
y = np.linspace(0, 5, 50)

# Create a mathematical function for conductivity
def conductivity(x, y):
  # Replace this with your actual conductivity equation
  # This example creates a bumpy surface with a peak in the center
  return (x**2 - 25) + (y**2 - 16)

# Generate conductivity data using the defined coordinates
X, Y = np.meshgrid(x, y)
Z = conductivity(X, Y)

# Create the contour plot
fig, ax = plt.subplots()
levels = np.linspace(Z.min(), Z.max(), 10)  # Adjust for desired number of levels
CS = ax.contour(X, Y, Z, levels=levels, cmap='plasma')  # Use a colormap for conductivity

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Conductivity (Condur) Plot')

# Add colorbar
plt.colorbar(CS, label='Conductivity')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Set the number of data points
num_points = 100

# Generate random angles between 0 and 2*pi (full circle)
theta = np.random.rand(num_points) * 2 * np.pi

# Generate random radii between 0 and 1
radius = np.random.rand(num_points)

# Create the polar plot
fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))
ax.scatter(theta, radius, s=50, c='blue', alpha=0.7, label='Random Points')

# Set labels and title
ax.set_xlabel('Theta (radians)')
ax.set_ylabel('Radius')
ax.set_title('Random Polar Plot')

# Add legend
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the first plot
x1 = np.linspace(0.0, 5.0, 100)
y1 = np.sin(2*np.pi*x1)

# Data for the second plot
x2 = np.linspace(0.0, 2.0, 50)
y2 = x2**2

# Create the figure and subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 5))  # 1 row, 2 columns

# Plot data on the first subplot
ax1.plot(x1, y1, label='Sine Wave')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Sine Function')
ax1.legend()

# Plot data on the second subplot
ax2.plot(x2, y2, label='Square Function')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Square Function')
ax2.legend()

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes = [20, 30, 25, 15, 10]
labels = ['Apple', 'Orange', 'Banana', 'Mango', 'Grapes']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen']
explode = (0.1, 0, 0, 0, 0)  # explode 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True)
plt.title('Pie Chart of Fruit Distribution')
plt.axis('equal')
plt.show()
```

```python
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2

plt.streamplot(X, Y, U, V, color='b', linewidth=1.5)
plt.title('Streamplot of Vector Field')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
