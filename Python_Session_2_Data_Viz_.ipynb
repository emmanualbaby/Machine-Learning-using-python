```python
import matplotlib.pyplot as plt


x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 2, 1]

# Plot lines
plt.plot(x, y)

# Add title and labels
plt.title("Simple Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show the plot
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 2, 1]

# Plot lines
plt.plot(x, y, color='green')

# Add title and labels
plt.title("Simple Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show the plot
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 2, 1]

# Create lists with shifted values
y_shifted_1 = [i + 1 for i in y]  # Add 1 to each element in y
y_shifted_2 = [i + 2 for i in y]  # Add 2 to each element in y
y_shifted_3 = [i + 3 for i in y]  # Add 3 to each element in y
y_shifted_4 = [i + 4 for i in y]  # Add 4 to each element in y

# Plot lines with different colors
plt.plot(x, y, color='r')  # Red (original y)
plt.plot(x, y_shifted_1, color='g')  # Green (shifted up by 1)
plt.plot(x, y_shifted_2, color='b')  # Blue (shifted up by 2)
plt.plot(x, y_shifted_3, color='k')  # Black (shifted up by 3)
plt.plot(x, y_shifted_4, color='c')  # Cyan (shifted up by 4)

# Add title and labels
plt.title("Simple Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show the plot
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 2, 1]

# Plot lines
plt.plot(x, y, color='green', linestyle="--")

# Add title and labels
plt.title("Simple Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show the plot
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 2, 1]

# Plot lines
plt.plot(x, y, 'o')

# Add title and labels
plt.title("Simple Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show the plot
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

plt.plot(x, y1 )
plt.plot(x, y2 )
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Lines Plot')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Lines Plot')
plt.legend()
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 10, 15, 20, 25]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Chart')
plt.show()
```

```python
data = [1, 1, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5]

plt.hist(data, bins=5)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
# Data for histogram
x = [1, 1, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5]

# Data for bar chart
y = [2, 4, 6, 8, 10]
categories = ['A', 'B', 'C', 'D', 'E']

# Create the histogram
plt.hist(x, bins=5, alpha=0.5, label='Histogram', color= "r")

# Create the bar chart
plt.bar(categories, y, alpha=0.5, label='Bar Chart')


# Add labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram vs. Bar Chart')

# Show the plot
plt.show()
```

```python
x = ['A', 'B', 'C', 'D']
y1 = [10, 15, 7, 10]
y2 = [5, 8, 10, 15]

plt.bar(x, y1, label='Group 1')
plt.bar(x, y2, bottom=y1, label='Group 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Stacked Bar Chart')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt


x = [1, 2, 3, 4, 5]
y1 = [2, 5, 7, 3, 1]
y2 = [1, 3, 2, 5, 6]

# Create the plot
plt.fill_between(x, y1, y2, color='skyblue', alpha=0.4)  # Fill between the two lines

# Customize the plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Filled Area Chart')
plt.grid(True)

# Add lines for reference
plt.plot(x, y1, label='Line 1', color='blue')
plt.plot(x, y2, label='Line 2', color='red')
plt.legend()

# Display the plot
plt.show()
```
